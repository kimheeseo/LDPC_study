```python
%(N,K) LDPC code에서 N은 code word, K는 message
k=input('enter the length of Msg Word:');
n=input('enter the length of codeword:');
P=input('enter the parity matrix :')

m=input('enter the Msg word:')
C=encode(m,n,k,'linear',G)
D=decode(C, n, k, 'linear',G)

H=[P' eye(n-k)]
dtable=syndtable(H)
#dtable은 decoding table을 나타내는 데, 이 값은 어떻게 구하느냐?
#S=CH^T를 통해 구할 수 있다. 만일 C(1*6)를 [0 1 0 0 0 0]이라고 하면,
#S의 값은 [1 1 0]이 되고, 이 값은 H의 2번째 열이므로 [0 1 0 0 0 0]에서
#2번째 element가 1로 0으로 바꾸면 된다.

#다시 한 번 예를 들면, 만일 C가 [1 0 0 1 0 0]이면, 1이 2개로 error가 2개인데,
#이 값도 위의 계산식처럼 S를 통해 진행하면, [0 1 1]이 되고, 이 값은 H의 5번째에
#위치하고 있지만, [1 0 0 1 0 0]은 1, 4번째 element에서 존재한다.

#따라서 H의 1, 4번째 값인 [1 1 1]+[1 0 0]=[0 1 1]의 값이 된다. 2개의 error를
#검출한 것을 확인할 수 있다.
```
