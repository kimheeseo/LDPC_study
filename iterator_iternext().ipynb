it.iternxt() 있는 경우

```python
import numpy as np

A = np.array([[10,20,30,40],[50,60,70,80]])

print("A값출력",A) #[[10,20,30,40],[50,60,70,80]]
print("A.shpae",A.shape,"\n") #(2,4)

it = np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])

while not it.finished: #이터레이터가 finished 위치가 아닐동안 반복
    idx = it.multi_index
    print("idx값",idx)
    print("A[idx]값",A[idx],"\n") #10부터 80까지 차례로 출력
    it.iternext() #이터레이터를 다음 위치로 넘기기
```

it.iternext() 없는 경우

```python
import numpy as np

A = np.array([[10,20,30,40],[50,60,70,80]])

print(A) #[[10,20,30,40],[50,60,70,80]]
print(A.shape) #(2,4)

it = np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])

while not it.finished: #이터레이터가 finished 위치가 아닐동안 반복
    idx = it.multi_index
    print(A[idx]) #10부터 80까지 차례로 출력

#it.iternext() 없는 경우
#idx값이 (0,0)에서 (0,1)로 update가 되지 않아서, 계속 10만 출력이 됩니다.
```
