```python
import numpy as np
y=np.exp(2)
print("y값(exp(2),",y)
```

```python
y=np.log(2)
print("y값(log(2),",y)
#아래의 값을 통해, "np.log"의 경우, 밑이 e(2.7)임을 알 수 있다.
```

```python
#L1=[0,0,1], P1=[0.1, 0.2, 0.7]
#cross_entropy
y=-(0*np.log(0.1)+0*np.log(0.2)+1*np.log(0.7))
print("손실함수로서의 cross_entropy값 :",y)
```

```python
#L1=[0,0,1], P1=[0.5, 0.3, 0.2]
#cross_entropy
y=-(0*np.log(0.5)+0*np.log(0.3)+1*np.log(0.2))
print("손실함수로서의 cross_entropy값 :",y)
```

```python
#entropy : 불확실성(uncertainty)에 대한 척도
#만약, 어떠한 사건이 같은 비율로 발생한다고 하고, 사건의 갯수를 n이라고 하면 entropy 는 log(n) 이다.
#-(1/n)*sigma[1*log(1/n)+0*log(1/n)]=log(n)

print("ex) 가방 안에 빨간공과 녹색 공이 50:50으로 들어있다면 \n")
print("공의 종류가 2개이기에, log(2)인 : ",np.log(2))

print("이때 가장 불확실성인(entropy)의 값이 큰데, 그 이유는 공이 같은 확률로 빨간색 아니면 녹색인데, 똑같은 확률이기에, 어떤 것이 나올지 모른다.")
```

```python
print("ex) 가방 안에 빨간공과 녹색공이 20:80 으로 들어있는 경우\n")
print("H(q)=-(0.2log(0.2)+0.8log(0.8))=",-(0.2*np.log(0.2)+0.8*np.log(0.8)))

print("같은 확률인, 가령 100개의 각기 다른 색깔과 모양의 공이 가방 안에 들어있다. \n")
print("가방에서 공을 꺼냈을 때 어떤 색과 모양을 갖는 공이 관찰될지 알기 매우 힘들다. \n")
print("entropy=log(n)=log(100)=",np.log(100),"\n")

print("100개의 공이 들어있다고 하더라도, 한 종류의 공이 99 %를 차지하고 있으면 어떨까?\n")
print("entropy는 매우 작아진다. -(0.99*log(0.99)+99*(0.01/99*log(0.01/99))) =", -(0.99*np.log(0.99)+0.01*np.log(0.01)))
```
