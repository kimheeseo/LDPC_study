# **Generator Matrix 생성**

```python
#출처 : https://youtu.be/M77dBHzlC3w
clc;
clear all;

%(N,K) LDPC code에서 N은 code word, K는 message
k=input('enter the length of Msg Word');
n=input('enter the length of codeword');
P=input('enter the parity matrix :')

G=[eye(k) P]
%G에서 eye(k)은 항등행렬, G인 생성행렬은 (K*N)행렬인데,
%G=[I_k P]이기 때문에, I_k를 항등행렬로 구성. 

%입력 : k=3, n=6, P=[1 1 1;1 1 0; 1 0 1]
```
